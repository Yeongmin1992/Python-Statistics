```python
def one_page(page):
        basic_url = "https://www.osulloc.com/kr/ko/shop/item/list?sort=review"
        url = basic_url + "&p={}".format(page - 1)
        response = requests.get(url)
        dom = BeautifulSoup(response.content, "html.parser")
        elements = dom.select(".list-wrapper.clearfix > .item")
        datas = []

        for element in elements:
            data = {
                "id": element.get("data-search-item-id"),
                "name": element.select_one(".name > a").text,
                "link": "https://www.osulloc.com" + element.select_one(".name > a").get("href"),
                "price": element.select_one(".price > strong").text,
                "img": "https://www.osulloc.com" + element.select_one(".defaultPic > img").get("src"),
            }
            datas.append(data)

        return pd.DataFrame(datas)
```
